```python
# Creamos un dataframe con los datos
df = pd.DataFrame(data=zip([-20,-10,0,10,20],
                           [-4,14,32,50,68]),
                  columns=['Datos', 'Resultados'])
df
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot the points
plt.scatter(df.Datos, df.Resultados, color='tab:blue', s=60)
plt.plot(df.Datos, df.Resultados, color='tab:orange')
plt.xlabel('Datos')
plt.ylabel('Resultados')
plt.show()
```

```python
X = df.Datos.to_numpy().reshape(-1, 1)
y = df.Resultados.to_numpy()
X
y
```

```python
# Almacena la data en variables
X = [[-20], [-10], [0], [10], [20]]
y = [-4, 14, 32, 50, 68]
```

```python
# Importa el modelo de RL
from sklearn import linear_model as lm
```

```python
# Ajusta el modelo a los datos
reg = lm.LinearRegression().fit(X, y)
```

```python
# Mide la exactitud del modelo
reg.score(X, y)
```

```python
# Devuelve y-interseccin
reg.intercept_
```

```python
# Devuelve la pendiente
reg.coef_
```

```python
# Predice el Resultado de
# un Dato desconocido
reg.predict([[55.4]])
```

```python
def celsius_to_fahrenheit(value):
  # y = 32 + 1.8x
  res = reg.intercept_ + reg.coef_[0]*value
  return res
```

```python
celsius_to_fahrenheit(55.4)
```
