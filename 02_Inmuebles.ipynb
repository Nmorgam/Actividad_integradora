# **Caso de Uso: Inmuebles**

## Datos

```python
import pandas as pd
```

```python
df = pd.read_csv('housedata.csv')
df
```

```python
df.info()
```

```python
df_num = df[df.columns[df.dtypes != 'object']]
df_num.head(3)
```

## Muestreo

```python
from sklearn.model_selection import train_test_split
```

```python
# 70% training, 30% testing
X_train, X_test, y_train, y_test = \
    train_test_split(
        df_num[df_num.columns[df_num.columns != 'price']],
    df_num['price'],
    test_size=0.3,
    random_state=42)
```

```python
X_test.head(5)
```

## Análisis

```python
df_copy = X_train.copy()
df_copy['price'] = y_train
corr = df_copy.corr()
corr.style.background_gradient(cmap='coolwarm')
```

## Ajuste

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

## Inferencia

```python
y_pred = reg.predict(X_test)
```

## Evaluación

```python
from sklearn.metrics import root_mean_squared_error
```

```python
rmse = root_mean_squared_error(y_test, y_pred)
print(f'RMSE: {rmse}')
```

## Mejora

```python
# Ajuste
reg_v2 = LinearRegression(fit_intercept=False, positive=True).fit(X_train, y_train)

# Inferencia
y_pred_v2 = reg_v2.predict(X_test)

# Evaluación
rmse_v2 = root_mean_squared_error(y_test, y_pred_v2)
print(f'RMSE: {rmse_v2}')
```

## Ingeniería de características

```python
df_new = df.drop(['street', 'country'], axis=1)

df_new['statezip'] = df_new['statezip'].str[3:].astype(int)

df_new['year'] = df_new['date'].str[:4].astype(int)
df_new['month'] = df_new['date'].str[5:7].astype(int)
df_new['day'] = df_new['date'].str[8:10].astype(int)
df_new.drop(['date'], axis=1, inplace=True)

df_new.head(3)
```

```python
df_new['yr_renovated'] = df_new['yr_renovated'].where(df_new['yr_renovated'] != 0, df_new['yr_built'])
df_new.head(3)
```

```python
# 70% training, 30% testing
X_train, X_test, y_train, y_test = \
    train_test_split(df_new[df_new.columns[df_new.columns != 'price']],
    df_new['price'],
    test_size=0.3,
    random_state=42)
```

#### ¿Cuál es el precio promedio de los inmuebles por ciudad?

```python
df_city_price = df[['price', 'city']].groupby('city').mean().reset_index().sort_values('price', ascending=False)
df_city_price.head(3)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure().set_figwidth(10)
plt.bar(df_city_price['city'], df_city_price['price'])
plt.ylabel('Price')
plt.title('Average Price by City')
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.preprocessing import TargetEncoder
import numpy as np
```

```python
encoder = TargetEncoder()
X_train['city'] = np.round(encoder.fit_transform(X_train['city'].to_numpy().reshape(-1,1), y_train),0).astype(int)
X_test['city'] = np.round(encoder.transform(X_test['city'].to_numpy().reshape(-1,1)),0).astype(int)
```

```python
X_test.head(3)
```

```python
# Ajuste
reg_v3 = LinearRegression(fit_intercept=False, positive=True).fit(X_train, y_train)

# Inferencia
y_pred_v3 = reg_v3.predict(X_test)

# Evaluación
rmse_v3 = root_mean_squared_error(y_test, y_pred_v3)
print(f'RMSE: {rmse_v3}')
```

```python
# Testing
plt.scatter(X_test['sqft_living'], y_test, s=5, label='testing')  # color='tab:blue'
plt.scatter(X_test['sqft_living'], y_pred_v3, s=5, label='prediction')  # color='tab:orange'
plt.xlim([500,4000])
plt.ylim([0.1,0.15e7])
plt.xlabel('sqft_living')
plt.ylabel('price')
plt.legend(loc='upper left')
plt.title(f'Testing - RMSE: {np.round(rmse_v3, 2)}')
plt.show()
```

```python
# Inferencia
y_train_pred = reg_v3.predict(X_train)

# Evaluación
rmse_train = root_mean_squared_error(y_train, y_train_pred)
print(f'RMSE: {rmse_train}')
```

```python
# Training
plt.scatter(X_train['sqft_living'], y_train, s=5, label='training')
plt.scatter(X_train['sqft_living'], y_train_pred, s=5, label='prediction')
plt.xlim([500,4000])
plt.ylim([0.1,0.15e7])
plt.xlabel('sqft_living')
plt.ylabel('price')
plt.legend(loc='upper left')
plt.title(f'Training - RMSE: {np.round(rmse_train, 2)}')
plt.show()
```
